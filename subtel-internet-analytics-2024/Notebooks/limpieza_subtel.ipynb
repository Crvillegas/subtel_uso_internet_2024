```python
import pandas as pd
import pyreadstat
```

```python
df_raw, meta = pyreadstat.read_sav(
    "../data/raw/subtel_decima_encuesta_2024.sav"
)
```

```python
columns_used = [
    "A2_EDAD_1",
    "A3_SEXO_1",
    "ZONA",
    "COD_REGION",
    "POND_USO"
]
df = df_raw[columns_used].copy()
```

```python
bins = [0, 17, 29, 44, 59, 100]
labels = ["0-17", "18-29", "30-44", "45-59", "60+"]

df["grupo_etario"] = pd.cut(
    df["A2_EDAD_1"],
    bins=bins,
    labels=labels
)

df["zona_desc"] = df["ZONA"].map({1.0: "Urbana", 2.0: "Rural"})
df["sexo_desc"] = df["A3_SEXO_1"].map({1.0: "Hombre", 2.0: "Mujer"})
```

```python
df_final = df.dropna()
df_final.shape
```

```python
df_final.to_csv(
    "../data/processed/internet_uso_zona_edad_chile.csv",
    index=False
)
```
